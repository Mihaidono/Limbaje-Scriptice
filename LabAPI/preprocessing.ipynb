```python
import pandas as pd
df = pd.read_csv('iris.csv')

df.head()
```

```python
import numpy as np
np.unique(df['variety'])
```

```python
df['class'] =df['variety'].replace({'Setosa':0, 'Versicolor':1, 'Virginica':2})
```

```python
df.head()
```

```python
df.describe()
```

```python
df = df.dropna() # stergem liniile care au coloane lipsa
```

```python
df = df.fillna(0) # completam valorile lipsa cu 0 sau cu o valoare anume
```

```python
df = df.drop_duplicates() # stergem liniile duplicate
```

```python
df['x'] = df[df.columns[:4]].values.tolist()
df['x']
```

```python
df['y'] = df['class'].apply(lambda x: x+1)
df['y']
```

```python
df.to_csv("iris_processed.csv",index=False)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=len(np.unique(df["y"])))
knn.fit(df["x"].tolist(), df["y"])
```

```python
y_pred = knn.predict([[5.3, 3.3, 5.7, 2.5]])

print(y_pred)
```

```python
from pyod.models.knn import KNN

clf_name = "KNN"
clf = KNN()
clf.fit(df["x"].tolist(), df["y"])
```

```python
clf.labels_
```

```python
clf.decision_scores_
```

```python
y_train_pred = clf.labels_  # binary labels (0: inliers, 1: outliers)
y_train_scores = clf.decision_scores_  # raw outlier scores

# get the prediction on the test data
y_test_pred = clf.predict([[5.5, 3.2, 1.1, 0.5]])  # outlier labels (0 or 1)
y_test_scores = clf.decision_function([[5.5, 3.2, 1.1, 0.5]])  # outlier scores

print(y_test_pred, "\n", y_test_scores)
```
